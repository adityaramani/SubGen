```python
from speech_recognition.speech_recognition import *
import pandas as pd
```

```python

rnn = RNNEngine()

def predict(cb,file_name):
    file_name, inf = cb(file_name)
    print(inf)
    return inf
```

```python
df = pd.read_csv("inferences.csv")
# df  = df.sample(frac=0.3)
```

```python
df['google_preds'] = df['wav_filename'].apply(gc_predict)
```

```python
df['google_preds']
```

```python
df['rnn_preds'] = df['wav_filename'].apply(lambda x : predict(rnn, x))
```

```python
df['sp_preds'] = df['wav_filename'].apply(lambda x : predict(sphinx.infer,x))
del(sphinx)
```

```python

from jiwer import wer

df['ds_wer'] = df.apply(lambda row : wer(row['transcript'], row['ds_preds']), axis = 1)
df['sp_wer'] = df.apply(lambda row : wer(row['transcript'], row['sp_preds']), axis = 1)
```

```python
from jiwer import wer
df['google_wer'] = df.apply(lambda row : wer(row['transcript'], row['google_preds'] if row['google_preds'] is  not None  else ''), axis = 1)
```
